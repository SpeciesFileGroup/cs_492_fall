```python
import matplotlib.pyplot as plt
```

```python
country_data_file = 'country_vs_names.log'
city_data_file = 'city_vs_names.log'
```

```python
with open('../data/' + country_data_file) as f:
    country_data = eval(f.read())
```

```python
country_data = country_data[:10]
countries = list(zip(*country_data))[0]
counts = list(zip(*country_data))[1]
x_pos = [i for i in range(len(countries))]

plt.bar(x_pos, counts, align = 'center')
plt.xticks(x_pos, countries)
plt.xticks(rotation=70)
plt.ylabel('Number of names')
plt.title('Countries vs names')
plt.grid()
plt.show()
```

```python
with open('../data/' + city_data_file) as f:
    city_data = eval(f.read())
```

```python
city_data = city_data[:10]
cities = list(zip(*city_data))[0]
counts = list(zip(*city_data))[1]
x_pos = [i for i in range(len(cities))]

plt.bar(x_pos, counts, align = 'center')
plt.xticks(x_pos, cities)
plt.xticks(rotation=70)
plt.ylabel('Number of names')
plt.title('Cities vs names')
plt.grid()
plt.show()
```
